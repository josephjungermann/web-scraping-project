```python
# import pandas
import pandas as pd
```

```python
# read csv in
df = pd.read_csv('../data/WARN_05012020.csv')
df
```

```python
# Dataset from 01/01/20 - 05/01/20 
# For updated totals, please use 'scrape.ipynb' in this folder to create/import new CSV
df.head()
```

```python
# reveal each column header
df.info()
```

```python
#fill all '-----' null values with '0' and correct any mistyped integers with correct values
df.affected[df.affected == '-----'] = 0
df.affected[df.affected == '(62'] = 62
```

```python
#convert 'affected' column from an object to an integer so we can add total employees laid off later
df['affected'] = df.affected.astype(int)
df.dtypes
```

```python
# find out what are the causes for why workers are being dislocated from their jobs
df.dislocation.value_counts()
```

```python
# group all Covid-19 related reasons for dislocation into new dataframe
df.loc[df['dislocation'].str.contains('COVID-19'), ['dislocation']] = 'COVID-19'
df
```

```python
# make sure previous code worked
df.dislocation.value_counts()
```

```python
df.business_type.value_counts()
```

```python
# create a sheet that filters out all other dislocation reasons and only includes COVID-19 related dislocation
covid_dislocation = "COVID-19"
```

```python
dislocation_list = df[df.dislocation == covid_dislocation]
```

```python
dislocation_list.head()
```

```python
dislocation_list.info()
```

```python
# find the top 20 industries that have the most entries in dataset for dislocation due to COVID-19 and sort values
dislocation_list.business_type.value_counts().reset_index().head(20)
```

```python
#Find top 20 companies that had the most people laid off in a single day
dislocation_list.sort_values("affected", ascending=False).head(20)
```

```python
#find the top 20 industries that have the greatest number of employees laid off from the start of the year to present
dislocation_list.groupby("business_type").affected.sum().reset_index().sort_values("affected", ascending = False).head(50)
```
